## Datatype Inconsistencies!
The provided data dictionary states that features Age, HoursPerWeek, and TotalHours should be integer datatypes. However, Pandas decided to designate these features as objects. This would indicate that there the values in the column are either str or mixed with numeric and non-numeric values.

Let's check the values in each of these columns.

```python
sc_player_data['Age'].unique() 
```

```python
sc_player_data['HoursPerWeek'].unique() 
```

```python
sc_player_data['TotalHours'].unique() 
```

As expected, there are str values in these columns '?', which represent missing values.

Let's now explore the rows with these '?' values.

```python
temp = sc_player_data[['LeagueIndex','Age','HoursPerWeek','TotalHours']]
```

```python
# How many rows have '?' in the Age column?
temp[temp['Age']=='?']
```

```python
# How many rows have '?' in the HoursPerWeek column?
temp[temp['HoursPerWeek']=='?']
```

```python
# How many rows have '?' in the TotalHours column?
temp[temp['TotalHours']=='?']
```

As the features HoursPerWeek and TotalHours have rows that contain '?' but inputed values in the other columns, it is important to be careful to note these edge cases and not assume that a '?' value automatically indicates that the other features will have a '?' as well.
